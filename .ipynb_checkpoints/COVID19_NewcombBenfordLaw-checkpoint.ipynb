### Reading data

```python
cases='csse_covid_19_time_series/time_series_covid19_confirmed_global.csv' #confirmed cases
deaths='csse_covid_19_time_series/time_series_covid19_deaths_global.csv' #deaths
data=cases
jhCsseURL='https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/'
frm=pd.read_csv(jhCsseURL+data,index_col=1)
frm=frm.transpose() #switching to have countries in columns

frmDict={}
for i in frm:
    r=list(frm[i].iloc[0:1])
    #r=float(frm[i][0])
    if len(r)>1:
        for q in range(len(r)):
            j=str(i)+':'+str(list(frm[i].iloc[0])[q])
            m=np.asanyarray(frm[i].iloc[3:]).T[q]
            frmDict[j]=m
    else:
        j=i
        k=frm[i].keys()[3:]
        m=frm[i][k]
        frmDict[j]=m
```

### Analysing Newcomb-Benford law

```python
def BenfordLawChecker(Data):
    ##visualizing digit profile
    v=abs(np.log10(Data))
    #print(v)
    vList=[]
    digNum=1 #digit number
    for i in v:
        digs=[]
        #print(i)
        if i==0:
            continue
        for j in str(i):
            if j!='0' and j!='.':
                k=str(i).index(j)
                break
        digs=str(i)[k:]
        digs=digs.replace('.','')
        #print(digs)
        vList.append(int(digs[digNum-1]))
    vList=np.array(vList)

    #frequency calculation
    freqDict={}
    for i in vList:
        if i in freqDict:
            freqDict[i]+=1
        else:
            freqDict[i]=1

    #probability calculation
    probDict={i:(freqDict[i])/sum(freqDict.values()) for i in sorted(freqDict.keys())}
    for i in np.linspace(1,9,9): #first digit
    #for i in np.linspace(0,9,10): #second digit
        if int(i) not in probDict:
            probDict[int(i)]=0

    #Benford's law
    xVals=np.linspace(1,9,9) #first digit
    yVals=np.log10(1+(1/xVals)) #first digit
    #xVals=np.linspace(0,9,10) #second digit
    #yVals=yVals=np.array([0.1197,0.1139,0.1088,0.1043,0.1003,0.0967,0.0934,0.0904,0.0876,0.0850]) #second digit

    #chisquare goodness of fit: sum of ((O-E)**2/E)
    chiSq,pVal=stats.chisquare(f_obs=list(probDict.values()),f_exp=yVals)
    print('Chisquare error: ',chiSq)
    print('Chisquare error pValue: ',pVal)

    #plot
    fig, ax = plt.subplots()
    fig=plt.figure()
    ax.bar(list(probDict.keys()),list(probDict.values()),color='y')
    ax.plot(xVals,yVals,'o',color='black')
    plt.show()
    
    return chiSq, pVal
```

```python
CountChiSqDict={}
CountpValDict={}
for i in frmDict:
    
    AllDatArray=np.array(list(frmDict[i]))
    AllDatArray=AllDatArray[AllDatArray!=0]

    if len(AllDatArray[AllDatArray!=0])<50:
        continue
    print(i)
    print('Non-zero entries: ',len(AllDatArray[AllDatArray!=0]))
    
    #sb.distplot(AllDatArray,color='b',hist=False)
    #sb.distplot(np.log10(AllDatArray),color='r',hist=False)
    
    #plt.title(i)
    #print(AllDatArray)
    print('Max Value: ',max(AllDatArray),'Min Value: ',min(AllDatArray))
    ChiSqErr,pVal=BenfordLawChecker(AllDatArray)
    CountChiSqDict[i]=ChiSqErr
    CountpValDict[i]=pVal
```

```python
response=requests.get('https://api.covidindiatracker.com/state_data.json')
CoVIndData=json.loads(response.text)
#CoVIndData=response.json()

#identify state code
StateIDict={ind+1:val['state'] for ind,val in enumerate(CoVIndData)}
#for key,value in StateIDict.items():
#    print(key,':',value)

#retrieve data for all states and districts
Cases='confirmed'
StatewiseDict={}
for i in StateIDict.keys():
    StateCode=i
    CaseList=[]
    for j in CoVIndData[StateCode-1]['districtData']:
        CaseList.append(j[Cases])
    StatewiseDict[StateIDict[i]]=CaseList
df=pd.DataFrame.from_dict(StatewiseDict,orient='index')
df=df.transpose()
AllData=[]
for i in StatewiseDict:
    for j in StatewiseDict[i]:
        AllData.append(j)
AllDatArray=np.array(AllData)

##visualizing distribution
sb.distplot(AllDatArray,color='b',hist=False)
sb.distplot(np.log10(AllDatArray),color='r',hist=False)

ChiSqErr,pVal=BenfordLawChecker(AllDatArray)
```

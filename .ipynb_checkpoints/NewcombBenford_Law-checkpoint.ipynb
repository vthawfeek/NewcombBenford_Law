### Benford's Law
When any natural data is transformed into its log10 form, the chances of occurance of the first digit follows a power law (as opposed to an expected uniform distribution) with lower digits appearing more times.<br/>
- This link contains useful mathematical details of the law: https://wwwf.imperial.ac.uk/~nadams/classificationgroup/Benfords-Law.pdf 
- This post contains a simple explanation of the concept: https://towardsdatascience.com/frawd-detection-using-benfords-law-python-code-9db8db474cf8
- This blog contains useful overview of different probability distributions: https://www.analyticsvidhya.com/blog/2017/09/6-probability-distributions-data-science/

### Imports

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
from  scipy import stats
```

### Continuous distributions

```python
#triangular distribution
##generate random numbers
n=int(1e5)
x=np.random.triangular(0,7,10,size=n)
sb.distplot(x,color='b')

sb.distplot(np.log10(x),color='r')
```

```python
#beta distribution
##generate random numbers
n=int(1e5)
x=np.random.beta(1,2,size=n)
sb.distplot(x,color='b')

sb.distplot(np.log10(x),color='r')
```

```python
#exponential distribution
##generate random numbers
n=int(1e5)
x=np.random.exponential(size=n)
sb.distplot(x,color='b')

sb.distplot(np.log10(x),color='r')
```

```python
#uniform distribution
##generate random numbers
n=int(1e5)
x=np.random.rand(n)
sb.distplot(x,color='b')

sb.distplot(np.log10(x),color='r')
```

```python
#normal distribution
##generate random numbers
n=int(1e5)
x = np.random.normal(loc=1,scale=0.05,size=n)
sb.distplot(x,color='b')

sb.distplot(np.log10(x),color='r')
```

### Comparison with Benford's law

```python
#visualizing digit profile
v=abs(np.log10(x))
vList=[]
digNum=1
for i in v:
    digs=[]
    for j in str(i):
        if j!='0' and j!='.':
            k=str(i).index(j)
            break
    digs=str(i)[k:]
    digs=digs.replace('.','')
    #print(digs)
    vList.append(int(digs[digNum-1]))
vList=np.array(vList)

#frequency calculation
freqDict={}
for i in vList:
    if i in freqDict:
        freqDict[i]+=1
    else:
        freqDict[i]=1

#probability calculation
probDict={i:(freqDict[i])/sum(freqDict.values()) for i in sorted(freqDict.keys())}

#Benford's law
xVals=np.linspace(1,9,9)
yVals=np.log10(1+(1/xVals))

#chisquare goodness of fit: sum of ((O-E)**2/E)
chiSq,pVal=stats.chisquare(list(probDict.values()),f_exp=yVals)
print('Chisquare error: ',chiSq)
print('Chisquare error pValue: ',pVal)

#plot
fig, ax = plt.subplots()
fig=plt.figure()
ax.bar(list(probDict.keys()),list(probDict.values()),color='y')
ax.plot(xVals,yVals,'o',color='black')
plt.show()
```

### Checking robustness

```python
def chiErrorCalculator(DistName,IterNum):
    n=IterNum
    if DistName=='normal':
        #x=np.random.normal(loc=1,scale=0.05,size=n)
        x=np.random.normal(size=n)
    elif DistName=='uniform':
        x=np.random.rand(n)
    elif DistName=='exponential':
        x=np.random.exponential(size=n)
    elif DistName=='beta':
        r1=1+np.random.randint(1,10)*np.random.rand()
        x=np.random.beta(1,r1,size=n)
    elif DistName=='triangular':
        r1=1+np.random.randint(1,10)*np.random.rand()
        r2=r1+np.random.randint(1,10)*np.random.rand()
        x=np.random.triangular(1,r1,r2,size=n)
    else:
        print('Distribution not found!')
    v=abs(np.log10(abs(x)))
    #print(v)
    vList=[]
    digNum=1
    for i in v:
        digs=[]
        for j in str(i):
            if j!='0' and j!='.':
                k=str(i).index(j)
                break
        digs=str(i)[k:]
        digs=digs.replace('.','')
        #print(digs)
        #print(len(digs))
        vList.append(int(digs[digNum-1]))
    vList=np.array(vList)
    #frequency calculation
    freqDict={}
    for i in vList:
        if i in freqDict:
            freqDict[i]+=1
        else:
            freqDict[i]=1
    #probability calculation        
    probDict={i:(freqDict[i])/sum(freqDict.values()) for i in sorted(freqDict.keys())}
    #chisquare goodness of fit: sum of ((O-E)**2/E)
    chiSq=stats.chisquare(list(probDict.values()),f_exp=yVals)[0]
    return chiSq


#Benford's law
xVals=np.linspace(1,9,9)
yVals=np.log10(1+(1/xVals))

reps=100
IterNum=int(1e5)
ChiDict={}
#for name in ['normal','uniform','exponential','beta','triangular']:
for name in ['normal','uniform','exponential']:
    print(name)
    DistName=name
    AvChiList=[]
    ChiErrList=[]
    for i in range(reps):
        ChiErrList.append(chiErrorCalculator(DistName,IterNum))
        AvChiVal=sum(ChiErrList)/len(ChiErrList)
        AvChiList.append(AvChiVal)
    ChiDict[DistName]=AvChiList
df=pd.DataFrame(ChiDict)
```

```python
fig, ax = plt.subplots()
fig=plt.figure()
for name in ['normal','uniform','exponential']:
    ax.plot(df[name],'-')
lab=['normal','uniform','exponential','beta','triangular']
ax.legend(lab,loc=0)
ax.set_title('All Distributions')
ax.set_xlabel('Cummulative iterations')
ax.set_ylabel('Average Chisquare residual')
plt.show()
```

### Manipulation effect

```python
n=int(1e5)
ManNum=int(1e3)
ManType='plus'
DistName='uniform'

if DistName=='normal':
    x=np.random.normal(loc=1,scale=0.05,size=n)
    #x=np.random.normal(size=n)
elif DistName=='uniform':
    x=np.random.rand(n)
elif DistName=='exponential':
    x=np.random.exponential(size=n)
else:
    print('No distribution found')
    
step=100
ChiSqDict={}
ran=np.linspace(0.001,0.9,step)
for deg in ran:
    ManDeg=deg
    for i in range(ManNum):
        ri=np.random.randint(n)
        if ManType=='plus':
            x[ri]=x[ri]+ManDeg*x[ri]
        elif ManType=='minus':
            x[ri]=x[ri]-ManDeg*x[ri]
        elif ManType=='multiply':
            x[ri]=x[ri]/ManDeg
        elif ManType=='divide':
            x[ri]=x[ri]*ManDeg
        else:
            print('Error!')

    #visualizing digit profile
    v=abs(np.log10(x))
    vList=[]
    digNum=1
    for i in v:
        digs=[]
        for j in str(i):
            if j!='0' and j!='.':
                k=str(i).index(j)
                break
        digs=str(i)[k:]
        digs=digs.replace('.','')
        #print(digs)
        vList.append(int(digs[digNum-1]))
    vList=np.array(vList)

    #frequency calculation
    freqDict={}
    for i in vList:
        if i in freqDict:
            freqDict[i]+=1
        else:
            freqDict[i]=1

    #probability calculation
    probDict={i:(freqDict[i])/sum(freqDict.values()) for i in sorted(freqDict.keys())}
    ChiSqDict[ManDeg]=list(probDict.values())

#Benford's law
xVals=np.linspace(1,9,9)
yVals=np.log10(1+(1/xVals))    

#plot
yData=list(ChiSqDict.values())
fig, ax = plt.subplots()
fig=plt.figure()
for num,val in enumerate(ChiSqDict):
    ax.bar(xVals+(0.00+(num+1)*(0.7)/step),yData[num],width=0.7/step)
lab=[str(i) for i in ran]
#ax.legend(lab,loc=0)
ax.set_title(ManType)
ax.plot(xVals,yVals,'o',color='black')
plt.show()
```
